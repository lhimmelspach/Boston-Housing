```python
"""
Supervised Learning Workflow:
Predicting House Prices
"""

### Problem Description

# The goal of this project is to build a predictive model that can estimate the price of a
# house based on its features. The target variable in this case we will be looking for is 
# house price. This will be a continuous variable.

# We will use a publicly available dataset containing information about the houses, such as
# size, number of bedrooms and bathrooms, location, and other features along with their sale
# price. I will be choosing a publicly available housing dataset that includes houses in the
# city of Boston. I'm choosing this particular city because it is the town I took a slightly
# expensive trip to run a marathon in once, and would like to see if it would cost less to
# purchase a house than make the same trip again. As someone who will also be looking to
# purchase a house in a different metro area, I think this one may make me feel better about
# the amount of money I spend on a house.
```

```python
# The learning task
# Using Regression analysis
# We will start by doing some Exploratory Data Analysis

# In this secion, we will load in the dataset and take an initial look to understand how it
# is set up, and potentially identify some issues and gain some insight that might be
# useful to begin the modeling process.

# Load the Libraries and Dataset
import pandas as pd
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Boston Housing dataset
boston = load_boston()
df = pd.DataFrame(boston.data, columns = boston.feature_names)
df['PRICE'] = boston.target

# Look at the first few rows
print("First 5 rows of the dataset:")
print(df.head())
print("\n")
```

```python
# Let us look at some basic info about the dataset

print("Dataset information:")
df.info()
print("\n")
```

```python
# We will also look at some descriptive statistics

print("Descriptive statistics:")
print(df.describe())
print("\n")
```

```python
# We'll print some visualizations of some features of the dataset

# Histogram of features
df.hist(figsize = (15, 10))
plt.suptitle("Histograms of Features", y = 1.02)
plt.tight_layout()
plt.show()

# Scatter plot of features vs price

for feature in boston.feature_names:
    plt.figure(figsize=(8,6))
    sns.scatterplot(x=df[feature], y = df['PRICE'])
    plt.title(f'{feature} vs Price')
    plt.xlabel(feature)
    plt.ylabel('Price')
    plt.show()

# Correlation heatmap

correlation_matrix = df.corr()
plt.figure(figsize=(12,10))
sns.heatmap(correlation_matrix, annot = True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
# Observations

# The distribution of the variable PRICE appears normal, but with a few outliers.
# Some of the features show linear relationships with the price, specifically RM
# which is the average number of rooms per house. 

# The correlation matrix helps show features that are strongly correlated with the
# price of the house and with other features. LSTAT shows a fairly strong negative
# correlation with the PRICE variable.
```

```python
# Starting Model Building and Training

# This section will prepare the data, select a model, train the model, and evaluate
# the model's performance.

# Preparing the data

# Data training/testing split
from sklearn.model_selection import train_test_split

X = df.drop('PRICE', axis = 1)
y = df['PRICE']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 5)
print("Shape of training features:", X_train.shape)
print("Shape of testing features:", X_test.shape)
print("Shape of training target:", y_train.shape)
print("Shape of testing target:", y_test.shape)
print("\n")
```

```python
# Scaling the features
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)

# Convert scaled data back to dataframes for easier inspection

X_train_scaled_df = pd.DataFrame(X_train_scaled, columns = X_train.columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns = X_test.columns)

print("First 5 rows of scaled training features:")
print(X_train_scaled_df.head())
print("\n")
```

```python
# Selecting and training the model

# Run a simple linear regression model
from sklearn.linear_model import LinearRegression

model = LinearRegression()

# Train the linear model on the scaled training data

model.fit(X_train_scaled, y_train)
```

```python
# Evaluating the model's performance on the test data
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

y_pred = model.predict(X_test_scaled)

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.3f}")
print(f"Mean Absolute Error: {mae:.3f}")
print(f"R-squared: {r2:.3f}")
print("\n")
```

```python
# Visualize the predictions against actual (test) prices

plt.figure(figsize = (8,6))
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color = 'red', linestyle='--')
plt.show()
```

```python
# Conclusion

# The R-squared value of 0.739 suggests that the model captures a significant portion
# of the variability in house prices, but there is still room for improvement.

# The mean squared error and mean absolute error provide a sense of the magnitude of the
# prediction errors in the original unit of the Price variable.

# Potential improvements could include creating new features from existing ones, such as
# polynomial features or interaction terms that might improve the model's ability to capture
# non-linear relationships.

# Additional complexity to the model, such as polynomial regression, decision tree-based
# models like Random Forest, or Gradient boosting could likely improve performance as well.

# Investigating some potential outliers in the data might improve the stability of the model
# and that could improve the generalization of the model.

# Identifying and using the more relevent features could simplify the model and also improve
# the performance.
```
